## 5.1 Script Ejemplo

En esta parte se desarrolla un ejemplo para entender la configuracion a partir de un script de python. En la figur se muestra la estructura del programa. El script empieza con una unidad de letura , en este caso VoltageReader. Luego, la cadena continuara con una unidad de procesamiento de Voltaje, que contiene 2 operaciones, una decodificacion e integracion incoherente. Luego una unidad de procesamiento es añadida. Dentro de la unidad de procesamiento, tenemos dos operaciones . Una es integracion incoherente y la otra es un ploteo de RTI para visualizar la data que esta siendo procesada. Finalmente, una unidad de escritura es creada para almacenar la data procesada.

El primer paso es establecer la ruta de búsqueda para los módulos. La ruta para las bibliotecas de python como numpy, matplotlib ya está configurada de manera predeterminada, por lo que solo necesitan importarse. Sin embargo, para usar también los módulos de Signal Chain, es necesario agregar la ruta del espacio de trabajo, que se realiza con las líneas de código a continuación.
